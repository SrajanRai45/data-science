```python
with open("data-files/massive_data.json","r") as file:
    data = json.load(file)

#data = json.loads(data)
data
```

```python
#pages you may like
#matrix format storage? i think yes due to weighted edges
def people_you_may_know(id,data):
    #making a list of friends
    friends = [user['friends'] for user in data['users'] if user['id'] == id]
    
    #cleaning list
    friends = friends[0]

    #same for pages
    liked_pages = [user['liked_pages'] for user in data['users'] if user['id'] == id]
    liked_pages = liked_pages[0]    #cleaning

    #pages you may like == pages = {name : score}
    pages = {}
    buffer = []    #for pages index storage
    
    #filling buffer
    for user in data['users']:
        if user['id'] in friends:
            buffer.extend(user['liked_pages'])  #extend removes square brackets [1,2,3]--> 1,2,3 and append all

    #filtering buffer of pages as you dont need to like the pages you have liked       
    buffer = list(filter(lambda x : x not in liked_pages,buffer))   

    #filling pages
    for page in data['pages']:
        if page['id'] in buffer:
            score = 0
            for i in buffer:
                if i == page['id']:
                    score += 1
            pages[page['name']] = score
    return pages
```

```python
user_id = 1
f = people_you_may_know(user_id,data)
f
```
