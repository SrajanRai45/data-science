```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

```python
evens = [i for i in range(1,1000) if i%2==0]
multiple_of_four = list(filter(lambda x:x%4==0,evens))
multiple_of_four
```
